```python
import time
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC 
from selenium.webdriver.common.action_chains import ActionChains
from pymongo import MongoClient
from bs4 import BeautifulSoup
import requests
import certifi
import re
import json


# Create the list of all indexes/currencies
def create_list(x,name):
    # Use certification
    ca = certifi.where()

    # Connect to the MongoDB server
    mongodb_client = MongoClient("mongodb+srv://mrinal:manimunnu@cluster0.40v6sv3.mongodb.net/?retryWrites=true&w=majority",tlsCAFile=certifi.where())

    # Select the database and collection
    database = mongodb_client["Final_Project_Data"]
    collection = database[name]
        
    data_list = []
    for item in x:
        
        index = item.find("a",{"data-test":"quoteLink"})
        # Get the url link for accessing the details of the index
        index_link = "https://finance.yahoo.com"+index['href']

        # Get the index symbol using regex remove"^"
        symbol =  re.sub(r'\^', '',index.text)
        # Get the name of index
        index_name = item.find("td",{"aria-label":"Name"}).text

        # Store the attributes in the MongoDB collection
        data = {"Index": symbol, "Name": index_name, "Link":index_link}
        data_list.append(data)

    # Insert the list of data into the collection
    collection.insert_many(data_list)

        
    

# Create a list in mongodb with basic info        
def create_list_world_indices(url,name):        
    # Read the url and header content
    header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'}

    # Get the response
    response = requests.get(url,headers=header)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Read all indexes
    search_result1 = soup.find_all("tr",{"class":"simpTblRow Bgc($hoverBgColor):h BdB Bdbc($seperatorColor) Bdbc($tableBorderBlue):h H(32px) Bgc($lv1BgColor)"})
    search_result2 = soup.find_all("tr",{"class":"simpTblRow Bgc($hoverBgColor):h BdB Bdbc($seperatorColor) Bdbc($tableBorderBlue):h H(32px) Bgc($lv2BgColor)"})
    
    # create the list of world indices
    create_list(search_result1,name)
    create_list(search_result2,name)        

    

# Using Selenium navigate historical 5Y daily data and save it to local drive
def save_historical_data(sub_url,sym):

    # Set up Selenium webdriver
    driver = webdriver.Chrome()
    time.sleep(5)

    # Navigate to URL
    driver.get(sub_url)    
    time.sleep(5)
    actions = ActionChains(driver)
    
    historical_data = driver.find_elements(By.XPATH,"//span[text()='Historical Data']")
    
    if len(historical_data)>0 :   
        #historical_data = driver.find_element(By.XPATH,"//span[text()='Historical Data']")
        historical_data[0].click()
        time.sleep(2)

        # Click on date range button
        date_range_button = driver.find_element(By.XPATH,"//div[@class='Pos(r) D(ib) C($linkColor) Cur(p)']")
        date_range_button.click()
        time.sleep(2)

        # Select 5 Year period
        five_year_data = driver.find_element(By.XPATH,"//span[text()='5Y']")
        five_year_data.click()
        time.sleep(2)
        actions.send_keys(Keys.ARROW_DOWN).perform()

        # Apply the changes
        apply = driver.find_element(By.XPATH,"//span[text()='Apply']")
        apply.click()
        time.sleep(10)
        
        # Scroll to the end of the page multiple times
        scroll_pause_time = 4 # adjust this value as per your need
        screen_height = driver.execute_script("return window.screen.height;")
        i = 1
        while True:
            # scroll to the end of the page
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            # wait for some time to let the page load
            time.sleep(scroll_pause_time)
            # calculate new scroll height and compare with previous scroll height
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == screen_height:
                # end of page is reached
                break
            else:
                screen_height = new_height
                i += 1
    

        # Save the resultant page in local drive
        filename = "historical_"+str(sym)+".html"
        #print(filename)
        with open(filename, 'w',encoding="utf-8") as f:
            f.write(driver.page_source)
    
        driver.quit()  
    else: 
        driver.quit()  
        
# Save the historical 5 Year data to local drive
def save_historical_local(name):

    # Use certification
    ca = certifi.where()

    # Connect to the MongoDB server
    mongodb_client = MongoClient("mongodb+srv://mrinal:manimunnu@cluster0.40v6sv3.mongodb.net/?retryWrites=true&w=majority",tlsCAFile=certifi.where())

    # Select the database and collection
    database = mongodb_client["Final_Project_Data"]
    collection = database[name]

    for i, document in enumerate(collection.find()):
        _id = document["_id"]
        symbol = document["Index"]
        url = document["Link"]

        # Call the historical data function to load and save the html file containing details
        save_historical_data(url,symbol)



# Read the historical data from local drive and save it to mongodb
def save_historicals_to_mongodb(file,sym,dbname):
    
    # Connect to the MongoDB server
    mongodb_client = MongoClient("mongodb+srv://mrinal:manimunnu@cluster0.40v6sv3.mongodb.net/?retryWrites=true&w=majority",tlsCAFile=certifi.where())

    # Select the database and collection
    database = mongodb_client["Final_Project_Data"]
    collection = database[dbname]
    
    # Check if the historicals page exists if not skip 
    if os.path.isfile(file):
        filename = file
        with open(filename, "r", encoding="utf-8") as f:
            html = f.read()

        # Parse the HTML with BeautifulSoup
        soup = BeautifulSoup(html, "html.parser")

        # Read the currency
        currency_detail = soup.find("div",{"class" : "C($tertiaryColor) Fz(12px)"}).text
        currency = re.search(r"Currency in (\w+)", currency_detail).group(1)

        # Read each daily price details
        all_blocks = soup.find_all("tr",{"class" : "BdT Bdc($seperatorColor) Ta(end) Fz(s) Whs(nw)"})

        # Check if the details are not present
        if all_blocks != [] :
            data_list = []
            for block in all_blocks:
                    date = block.find("td",{"class":"Py(10px) Ta(start) Pend(10px)"}).text
                    other_details = block.find_all("td",{"class":"Py(10px) Pstart(10px)"})
                    open_price = other_details[0].text
                    high_price = other_details[1].text
                    low_price = other_details[2].text
                    close_price = other_details[3].text
                    adj_close_price = other_details[4].text
                    volume = other_details[5].text

                    data = {"Index" : sym ,"Currency" : currency ,"Date": date, "Open": open_price, "High":high_price, "Close" : close_price, "Adj. Close" :adj_close_price, "Volume" : volume}
                    data_list.append(data)

            # Insert the list of data into the collection
            collection.insert_many(data_list)

        f.close()
    
# Read each index symbol and retrieve the file
def read_local_historicals(list_name,dbname):

    # Use certification
    ca = certifi.where()

    # Connect to the MongoDB server
    mongodb_client = MongoClient("mongodb+srv://mrinal:manimunnu@cluster0.40v6sv3.mongodb.net/?retryWrites=true&w=majority",tlsCAFile=certifi.where())

    # Select the database and collection
    database = mongodb_client["Final_Project_Data"]
    collection = database[list_name]

    for i, document in enumerate(collection.find()):
        _id = document["_id"]
        symbol = document["Index"]
        filename = "historical_"+str(symbol)+".html"
        save_historicals_to_mongodb(filename,symbol,dbname)

        
def world_indices():
    # Create a list of world_indices
    world_indices_url = "https://finance.yahoo.com/world-indices"
    create_list_world_indices(world_indices_url,"List_of_World_Indices")  

    # Save the historical pages locally for world indices
    save_historical_local("List_of_World_Indices")

    # Save the historical data of currency, prices and volume in mongodb
    read_local_historicals("List_of_World_Indices","World_Indices_Price_and_Volume_5Y")
    
    
def world_currencies():
    # Create a list of currencies
    currencies_url = "https://finance.yahoo.com/currencies"
    create_list_world_indices(currencies_url,"List_of_Currencies") 

    # Save the historical pages locally for currencies
    save_historical_local("List_of_Currencies")

    # Save the historical data of currency, prices and volume in mongodb
    read_local_historicals("List_of_Currencies","World_Currencies_Price_and_Volume_5Y")

    
# Run the code for generating list of world indices and the 5Y historical data to mongodb
world_indices()

# Run the code for generating list of world currencies and the 5Y historical data to mongodb
world_currencies()
```
